```python
# Import packages
import FinanceDataReader as fdr
import yfinance as yf
import pandas as pd
```

```python
symbols = []
df_2023 = []
df_2022 = []
df_2021 = []
market_cap = []
error = []

for i, symbol in enumerate(fdr.StockListing('S&P500')['Symbol']):
    if i == 213: # ERROR
        continue
    try:
        financials = yf.Ticker(symbol).financials.T
        info = yf.Ticker(symbol).info
        if 'marketCap' in info and info['marketCap'] is not None:
            symbols.append(symbol)
            df_2023.append(financials.iloc[1])
            df_2022.append(financials.iloc[2])
            df_2021.append(financials.iloc[3])
            market_cap.append(info['marketCap'])
            print(f"i: {i}, {len(market_cap), len(df_2023)}")
        else:
            print(symbol)
    except:
        error.append((i, symbol))


symbols = pd.DataFrame(symbols)
df_2023 = pd.DataFrame(df_2023)
df_2022 = pd.DataFrame(df_2022)
df_2021 = pd.DataFrame(df_2021)
market_cap = pd.DataFrame(market_cap)
```

```python
cols_to_drop = df_2023.columns[df_2023.isna().sum() >= 100]

df_2023 = df_2023.drop(columns=cols_to_drop)
df_2022 = df_2022.drop(columns=cols_to_drop)
df_2021 = df_2021.drop(columns=cols_to_drop)
```

```python
df_2023 = df_2023.set_index(symbols.iloc[:, 0])
df_2022 = df_2022.set_index(symbols.iloc[:, 0])
df_2021 = df_2021.set_index(symbols.iloc[:, 0])
```

```python
market_cap.columns = ['Market Cap']
market_cap = market_cap.set_index(symbols.iloc[:, 0])
```

```python
df_2023 = pd.concat([df_2023, market_cap], axis=1)
df_2022 = pd.concat([df_2022, market_cap], axis=1)
df_2021 = pd.concat([df_2021, market_cap], axis=1)
```

```python
def fill_na(df):
    for col in df.columns:
        if col != 'Market Cap':
            non_missing_data = df[['Market Cap', col]].dropna()
            # import pdb; pdb.set_trace()
            average_ratio = (non_missing_data[col].sum()) / (non_missing_data['Market Cap'].sum())
            print(average_ratio)
            df.loc[df[col].isna(), col] = df.loc[df[col].isna(), 'Market Cap'] * average_ratio

fill_na(df_2023)
fill_na(df_2022)
fill_na(df_2021)
```

```python
def convert_percent_to_float(df, column_list):
    for column in column_list:
        # Check if any value in the column contains a '%'
        if df[column].dtype == object and df[column].str.contains('%').any():
            # Remove '%' sign and convert to float
            df[column] = df[column].str.replace('%', '').astype(float) / 100
            print(column)

# Apply the conversion function
convert_percent_to_float(df_2023, df_2023.columns)
convert_percent_to_float(df_2022, df_2022.columns)
convert_percent_to_float(df_2021, df_2021.columns)
```

```python
log_transform_cols = df_2023.columns[(df_2023 > 1e+4).any()]
print(log_transform_cols)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
import numpy as np

def signed_log1p(x):
    return np.sign(x) * np.log1p(np.abs(x))

df_2023[log_transform_cols] = df_2023[log_transform_cols].apply(signed_log1p)
df_2022[log_transform_cols] = df_2022[log_transform_cols].apply(signed_log1p)
df_2021[log_transform_cols] = df_2021[log_transform_cols].apply(signed_log1p)
```

```python
scaler = StandardScaler()
df_2023[df_2023.columns] = scaler.fit_transform(df_2023)
df_2022[df_2022.columns] = scaler.fit_transform(df_2022)
df_2021[df_2021.columns] = scaler.fit_transform(df_2021)
```
