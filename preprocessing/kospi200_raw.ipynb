```python
import FinanceDataReader as fdr
from pykrx import stock
```

```python
kospi200_codes = stock.get_index_portfolio_deposit_file("1028")
tickers = [symbol.zfill(6) for symbol in kospi200_codes]
```

```python
kospi_base = fdr.StockListing('KOSPI') # KOSPI: 940 종목
kospi_base
```

```python
mask = kospi_base['Code'].isin(kospi200_codes)
kospi200_base = kospi_base[mask]
kospi200_base
```

```python
kospi_add = stock.get_market_fundamental("20240529", market="KOSPI")
kospi_add
```

```python
mask2 = kospi_add.index.isin(kospi200_codes)
kospi200_add = kospi_add[mask2]
kospi200_add
```

```python
kospi200_add.reset_index(inplace=True)

kospi200_add.rename(columns={'티커': 'Code'}, inplace=True)

print(kospi200_add)
```

```python
kospi200 = kospi200_base.merge(kospi200_add, on='Code', how='inner')
kospi200
```

```python
kospi200 = kospi200.drop(['Open', 'High', 'Low', 'Volume'], axis=1)
kospi200
```

```python
kospi200 = kospi200.drop(['Market', 'Dept', 'ISU_CD', 'MarketId'], axis=1)
kospi200
```
