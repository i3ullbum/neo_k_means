```python
df = df.drop(['% Change'], axis=1)
```

```python
df = df.drop(['FCF Growth', 'Net WC'], axis=1)
```

```python
df = df.drop(['EPS Growth'], axis=1)
```

```python
print(df.isna().sum())
```

```python

non_missing_data = df.dropna()
non_missing_data['PE Ratio'].sum()
```

```python
df['Debt Growth (YoY)'] = df['Debt Growth (YoY)'].astype(str)
df['Debt Growth (YoY)'] = df['Debt Growth (YoY)'].str.replace('%', '', regex=False).astype(float) / 100
```

```python
# Assume values are proportional to market cap
column_to_fill = [
    'PE Ratio',
    'Income Tax',
    'Total Debt',
    'Debt Growth (YoY)',
    'PB Ratio',
    'EV/EBITDA',
    'Beta (1Y)',
    'ATR',
]

for col in column_to_fill:
    df[col] = pd.to_numeric(df[col], errors='coerce')

    non_missing_data = df[['Market Cap', col]].dropna()

    average_ratio = float(non_missing_data[col].sum()) / float(non_missing_data['Market Cap'].sum())
    df.loc[df[col].isna(), col] = df.loc[df[col].isna(), 'Market Cap'] * average_ratio
```

```python
print(df.isna().sum().sum())
```

```python
df.to_csv('sp500_fillna.csv')
df
```
